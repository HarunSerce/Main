```python

# Function to ask the user to input something for the story
def story_add(name):
    vowels = ['i', 'o', 'u', 'a', 'e']
    first_letter = name[0]
    if first_letter in vowels:
        question = f'Enter an {name}: '
    else:
        question = f'Enter a {name}: '

    while True:
        story_item = input(question)
        if story_item.isalpha():
            return story_item.lower()
        else:
            print('Please enter a word')

# Function to generate the first funny Roman story
def generate_funny_roman_story1():
    name1 = story_add('adjective')
    name2 = story_add('noun')
    name3 = story_add('verb')
    name4 = story_add('adverb')
    name5 = story_add('emotion')
    name6 = story_add('noun')

    story = f'In ancient Rome, there were two friends, {name1} and {name2}, known for their {name3} antics. One day, they decided to {name4} to a {name5} gladiator competition. When they arrived, they {name3} their way into the arena and challenged the fiercest gladiator to a {name1} duel. To everyone\'s surprise, {name1} and {name2} used {name6} and {name6} as their weapons. The crowd roared with {name5}, and the emperor declared them the {name1} champions of Rome.'

    return story

# Function to generate the second funny Roman story
def generate_funny_roman_story2():
    name1 = story_add('adjective')
    name2 = story_add('noun')
    name3 = story_add('verb')
    name4 = story_add('adverb')
    name5 = story_add('emotion')
    name6 = story_add('noun')

    story = f'In ancient Rome, a famous philosopher named {name1} decided to {name3} a {name2} with his {name6}. This {name2} was known for its {name4} wisdom and its ability to {name3} people. As {name1} approached the {name2}, he {name3} and exclaimed, "{name5}!" The {name2} simply {name3} and replied, "{name4}, my friend." Everyone in Rome laughed at the {name1} and his comical encounter with the {name2}.'

    return story

# Function to generate the third funny Roman story
def generate_funny_roman_story3():
    name1 = story_add('adjective')
    name2 = story_add('noun')
    name3 = story_add('verb')
    name4 = story_add('adverb')
    name5 = story_add('emotion')
    name6 = story_add('noun')

    story = f'In ancient Rome, the mischievous twins, {name1} and {name2}, were notorious for their {name3} pranks. One day, they decided to {name3} the grand {name2} of Emperor {name6}. With {name4} stealth, they sneaked into the palace and {name3} the emperor\'s {name2} with {name1}. The emperor awoke with a start and shouted, "{name5}!" The twins fled, leaving the emperor in a state of {name5} and the entire palace in {name5} chaos.'

    return story

# Function to select a story based on user input
def story_selector():
    while True:
        story_select = input('Enter a number between one and three to select a story: ').lower()
        if story_select in ('one', 'two', 'three'):
            return {'one': 1, 'two': 2, 'three': 3}[story_select]
        elif story_select.isnumeric():
            int_story_select = int(story_select)
            if 1 <= int_story_select <= 3:
                return int_story_select
        print('Please enter a valid number between one and three.')

# Function to ask if the user wants to play again
def play_again():
    while True:
        play_choice = input('Do you want to play again? (yes/no): ').lower()
        if play_choice in ('yes', 'no'):
            return play_choice
        print('Please enter "yes" or "no".')

# Greeting
print('Welcome to the Funny Roman Stories Mad Libs Game!')

# Define a dictionary to map story numbers to story generator functions
story_generators = {
    1: generate_funny_roman_story1,
    2: generate_funny_roman_story2,
    3: generate_funny_roman_story3
}

# Main game loop
while True:
    selected_story = story_selector()

    # Check if the selected_story is a valid key in the story_generators dictionary
    if selected_story in story_generators:
        generated_story = story_generators[selected_story]()
        print('\nHere\'s your funny Roman story:\n')
        print(generated_story)
    else:
        print('Invalid story selection.')

    play_choice = play_again()
    if play_choice == 'no':
        print('Thanks for playing! Goodbye!')
        break
```
