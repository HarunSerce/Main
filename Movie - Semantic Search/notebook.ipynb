```python
import pandas as pd
```

```python
df = pd.read_csv('imdb.csv')
```

```python
df['Cast'] = df['Star1'] + ',' +df['Star2'] + ',' + df['Star3'] + ',' +df['Star4'] + ','
```

```python
df.columns
```

```python
columns = ['Certificate','Meta_score','Star1', 'Star2', 'Star3', 'Star4','Gross']
```

```python
df= df.drop(columns=columns,axis=1)
```

```python
df.columns
```

```python

df = df[['Poster_Link', 'Series_Title', 'Released_Year', 'Runtime', 'Genre','IMDB_Rating', 'Overview', 'Director','Cast','No_of_Votes']]
```

```python
df.head()
```

```python
df = df.rename(columns = {'Poster_Link':'Poster', 'Series_Title':'Title', 'Released_Year':'Year', 'Runtime':'Duration','IMDB_Rating':'Rating', 'No_of_Votes':'Votes'})
```

```python
df.head()
```

```python
df['Duration'] = df['Duration'].str.extract('(\d+)').astype(int)
```

```python
df.head()
```

```python
df.to_csv('Movieset.csv')
```
